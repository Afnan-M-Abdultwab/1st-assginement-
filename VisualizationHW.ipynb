```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('50ulke.csv')
```

```python
del file['Unnamed: 0']
file
```

```python
file = pd.DataFrame(file)
file.groupby('Continent').max()
```

This table shows the maximum national income , growth and per person in each Continent and to which country does it belong.

```python
file.info()
```

There are 50 rows in 5 columns starting from number zero to 49 and all of them have full information ; no null values

```python
file.describe()
```

```python
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.5)
sns.boxplot(x="Continent", y="National Income", data=file)
```

Here it shows that the least Continent with national income range is Oceania 
& the biggest range belongs to North America up to 20,000 yet the density of the national income is not clear enough.

```python
sns.violinplot(x="Continent", y="National Income", data=file)
sns.stripplot(x="Continent", y="National Income",size=7, data=file)
```

Now this polt explained how Asia & Europe got the most densities within them even though it goes around specific range but it is important because it has the largest numbers of people, while on the other hand North Amercia has only little density in the higher range.

```python
file.corr()
```

```python
sns.heatmap(file.corr() , annot=True)
```

```python
sns.lmplot(x='Per person', y='National Income', data=file ,col='Continent',aspect=0.9, height=5)
```

This polt explains the slope between national income and it's effect per person in each Continent
almost all the Coninent have no slope but a fixed range per person while in North America it has linear positive regression slope.

```python
fig, ax = plt.subplots()
sns.lineplot(x='Growth', y='National Income', data=file, alpha=1, dashes=False, markers=True, )
ax.tick_params(axis = 'both', which = 'major')
ax.tick_params(axis = 'both', which = 'minor')

ax.set_xlabel('Growth')
ax.set_ylabel('National Income')
```

This plot proves that there is not a certain correlaton betweeb growth and national incomes ; growth is not the most important feature for a national income then , it must depend on other factors.

```python
top = file[file['National Income'] >=1646.34]
top
```

This table shows the Countries with National Income above the mean , Europe has 5 counties above the mean followed by Asia with 4 counties and North America only got 2 Counties above the mean.

```python
sns.lmplot(x='Growth', y='National Income', data=top[0 : 3] , col='Countries')
```

```python
sns.lmplot(x='Growth', y='National Income', data=top[3 : 6] , col='Countries')
```

```python
sns.lmplot(x='Growth', y='National Income', data=top[6 : 9] , col='Countries')
```

```python
sns.lmplot(x='Growth', y='National Income', data=top[9:10] , col='Countries')
```

These plots expose the correlation between growth and National Income in the  top 10 Countries

```python
sns.barplot(x='Per person', y='Countries',data = top)
```

The income per person for the top countries ; USA , Japan and Canda are the highest when it comes to income per person.
